```python
# Map 만들기

from NeuroMap import Map, Cell, Arrow
from torch.nn import Linear, Softmax
from NeuroMap.soma import Test

L1 = Linear(10, 20)
L2 = Linear(20, 30)
SM = Softmax(dim=1)
L3 = Linear(30, 5)
DT = Test()

model = Map([
    [Cell(L1, Arrow.RIGHT), Cell(L2, Arrow.DOWN)],
    [Cell(L3, Arrow.STOP),  Cell(DT, Arrow.LEFT)]
    ])
```

```python
# 손실함수, 옵티마이저 설정

import torch.nn as nn
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
# 랜덤 데이터 생성
# 100개의 샘플, 입력 크기 10, 클래스 5개짜리 문제

import torch

X = torch.randn(100, 10)
y = torch.randint(0, 5, (100,))
```

```python
# 학습 루프

epochs = 30
for epoch in range(epochs):

    optimizer.zero_grad()

    # NeuroMap은 입력을 던지면 자동으로 Arrow를 따라 흐름을 계산함
    outputs = model(X)

    loss = criterion(outputs, y)
    loss.backward()
    optimizer.step()

    print(f"[Epoch {epoch+1}/{epochs}] Loss = {loss.item():.4f}")

print("학습 완료!")
```
